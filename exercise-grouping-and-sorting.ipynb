## 3.
What are the minimum and maximum prices for each `variety` of wine? Create a `DataFrame` whose index is the `variety` category from the dataset and whose values are the `min` and `max` values thereof.

```python
reviews.groupby(['variety']).price.agg([len, 'min', 'max'])
```

```python
# From https://stackoverflow.com/questions/12589481/multiple-aggregations-of-the-same-column-using-pandas-groupby-agg
# df.groupby('kind').agg(
#     max_height=('height', 'max'), min_weight=('weight', 'min'),)

reviews.groupby('variety').agg(
    Min_price=('price','min'), Max_price =('price', 'max')
)
reviews.groupby('variety').agg(
    Min_price =pd.NamedAgg(column='price', aggfunc='min'),
    Max_price =pd.NamedAgg(column='price', aggfunc='max')
)
```

```python
# dito from https://stackoverflow.com/questions/12589481/multiple-aggregations-of-the-same-column-using-pandas-groupby-agg
# df.groupby('dummy').agg(Mean=('returns', 'mean'), Sum=('returns', 'sum'))
reviews.groupby(['variety']).agg(Mean =('price', 'min'), Max=('price', 'max'))
```
